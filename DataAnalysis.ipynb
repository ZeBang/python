### Step 3: Plot the data (3 points)

Plotting is very useful to help you understand your dataset.

Use `matplotlib` library to create basic Bar, Histogram, Scatter, Pie plots.

For example, the following histogram shows the distribution of gross millions.



````python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(6, 3))
plt.hist(movies_df['gross_millions'], bins=50)
plt.title('Distribution of Gross Revenue for Top-Grossing Movies')
plt.xlabel('Gross Revenue (Millions of Dollars)')
plt.ylabel('Number of Movies')
plt.show()
````
